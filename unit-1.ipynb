### Law of Increasing Costs

As production of one good increases, the opportunity cost of producing additional units of that good also increases. **This is because resources are not perfectly adaptable to the production of all goods.** As more resources are allocated to produce one good, less efficient resources must be used, leading to higher opportunity costs.

This is the reason why the production possibilities frontier (PPF) is typically bowed outward instead of a theoretical straight line. The bowed shape reflects the increasing opportunity costs associated with reallocating resources from one good to another.

```python
x_theo = np.linspace(0, 10, 81)
y_theo = 81 - 9 * x_theo

x_real = np.linspace(0, 10, 81)
y_real = 81 - x_real**2

plt.figure(figsize=(8, 6))
plt.plot(x_theo, y_theo, label="Theoretical PPF", linestyle="--", color="black")
plt.plot(x_real, y_real, label="Real PPF", color="black")
plt.text(
    0.2,
    4,
    "As production of Good A increases,\nthe opportunity cost of Good B increases.",
    fontsize=12,
    bbox=dict(facecolor="white", alpha=0.5, edgecolor="black"),
)
plt.title("Law of Increasing Opportunity Costs")
plt.xlabel("Good A")
plt.ylabel("Good B")
plt.grid(True)
plt.legend()
plt.xlim(0, 10)
plt.ylim(0, 100)

plt.show()
```
